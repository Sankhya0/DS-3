```python
# -*- coding: utf-8 -*-
"""
Created on Mon Aug 29 14:11:07 2022

@author: aman mohanty
"""
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("pima-indians-diabetes.csv")
atts=['pregs','plas','pres','skin','test','BMI','pedi','Age']
# Q1
def stats(D1):
    
    print("Mean of "+str(D1)+ " is "+str(statistics.mean(data[D1])))
    print("Median of "+str(D1)+ " is "+str(statistics.median(data[D1])))
    print("Mode of "+str(D1)+ " is "+str(statistics.mode(data[D1])))
    print("Minimum of "+str(D1)+ " is "+str(min(data[D1])))
    print("Maximum of "+str(D1)+ " is "+str(max(data[D1])))
    print("Standard Deviation of "+str(D1)+ " is "+str(statistics.stdev(data[D1])))
    print()



for i in atts:
    stats(i)

# Q2
def scaplt(a):
    if (a!="Age"):
        x=data["Age"]
        y=data[a]
        plt.scatter(x,y)
        plt.xlabel("Age",fontsize=10)
        plt.ylabel(a,fontsize=10)
        plt.grid()
        plt.show()

def scaplt1(a):
    if (a!="BMI"):
        x=data["BMI"]
        y=data[a]
        plt.scatter(x,y)
        plt.xlabel("BMI",fontsize=10)
        plt.ylabel(a,fontsize=10)
        plt.grid()
        plt.show()


    
for i in atts:
    scaplt(i)
    
for i in atts:
    scaplt1(i)

# Q3
def corrage(a):
    if(a!="Age"):
        print("Correlation of Age with "+str(a)+" is "+str(data["Age"].corr(data[a])))
        

def corrBMI(a):
    if(a!="BMI"):
        print("Correlation of BMI with "+str(a)+" is "+str(data["BMI"].corr(data[a])))

for i in atts:
    corrage(i)
print()
for i in atts:
    corrBMI(i)

# Q4
plt.xlabel('Number of times pregnant')  
plt.ylabel('Number of times')  
plt.hist(data["pregs"],17)
plt.grid()
plt.show()

plt.hist(data["skin"],20)
plt.xlabel('Tricep fold thickness(mm)')
plt.ylabel('Number of times')
plt.grid()
plt.show()

# Q5
preg1=[]
preg0=[]
l=len(data["pregs"])

for i in range(l):
    if(data["class"][i]==0):
        preg0.append(data["pregs"][i])
    else:
        preg1.append(data["pregs"][i])

plt.hist(preg0)
plt.xlabel('Number of times pregnant')
plt.title('CLASS 0')
plt.grid()
plt.show()

plt.hist(preg1)
plt.xlabel('Number of times pregnant')
plt.title('CLASS 1')
plt.grid()
plt.show()
    
# Q6
def boxplot(a):
    if(a!="class"):
        plt.title("Boxplot of "+str(a))
        plt.boxplot(data[a])
        plt.grid()
        plt.show()

for i in atts:
    boxplot(i)

    
    

    

```
